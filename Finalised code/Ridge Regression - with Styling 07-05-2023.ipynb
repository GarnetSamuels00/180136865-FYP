The function "backtest" does the work of making the predictions because i am using time series data I need to make sure that i am using historical data to predict future data. This function splits the data up by seasons and uses the past seasons to predict the future seasons. As the start parameter is set to 2 this function will need at least 2 seasons of data to start predicting future seasons. For example 2016-17 data would be used to predict 2018 data. Then 2016-18 data will be used to predict 2019 data and so on

If you wanted to make predictions for two seasons at a time you would set the step parameter to 2.


I then fit my model which takes the training data so all of my predictors which is what i am actually giving the model to make its judgement. Then there is the target which is the actual thing that I am trying to predict this is fed into the model so it can figure out which predictors correlate to the target.

Making predictions on the same data I train on will make the model look really good but its unrealistic.

```python
sorted(dataFrame["season"].unique())
```

```python
def backtest(data, model, predictors, start=2, step=1):
    #a list of data frames where each data frame is a 
    #prediction for a single season.
    all_predictions = []
    
    #seasons will be a list of all the seasons that exist in the data
    seasons = sorted(data["season"].unique())
    
    #used to loop through the seasons
    #start parameters default value of 2 can be overrun when you call the backtest function
    #len will be the length of the seasons 
    #the step parameter is used to progress through seasons
    for i in range(start, len(seasons), step):
        #pulls out the season number
        season = seasons[i]
        
        #splits up the training data from the test data
        #training data is used to train my machine learning algorithm which 
        #will use all of the data that comes before the current season 
        train =data[data["season"] < season]
        #the test data is what I will use to generate predictions which 
        #is going to be on the current season
        test = data[data["season"] == season]
        
        #fit model with training data and predictions which is what 
        #the model will use to make its judgements
        #the target is the actual thing I am trying to predict
        #the target is used to figuure out which predictors actually correalte to the target
        model.fit(train[predictors], train["target"])
        
        #generate my predictions by making predictions on the test set
        #by default returns a numPy array
        preds = model.predict(test[predictors])
        #converts numPy array into a Pandas Series
        #I keep the index the same as the test set index
        #this makes it easier to combine the actuals and predictions
        preds = pd.Series(preds, index=test.index)
        
        #creates a Data Frame called combined
        #the pandas concatonate function which puts two series or data frames together
        #the actual values the target is put together with the predictions
        #axis=1 will treat target and predictions as seperate columns thus making 2 columns
        combined = pd.concat([test["target"], preds], axis=1)
        #renaming the two columns as "actual" and "prediction"
        combined.columns = ["actual", "prediction"]
        
        #append the combined data frame to the list all_predictions 
        #all predictions is going to be a list of 5 different Data Frames
        all_predictions.append(combined)
        #combines all 5 data frames into one 
        #pass in the axis parameter 
        #all these data frames have the same column but we want to combine them as seperate rows 
    return pd.concat(all_predictions)
```

Rolling Averages - The team could have had a random game where the team plays differently for whatever reason. Taking an average of how the team performed in the last x amount of games and use that to give my algorithm information it can use to predict how the team will perform in their next game 

```python
dataFrame
```

```python
#data frame called df-rolling
#selected_columns is added into the data frame
#the extra columns "won", "team" and "season" are added to compute the rolling averages
df_rolling = dataFrame[list(selected_columns) + ["won","team","season"]]
```

```python
df_rolling
```

```python
#function to find the team averages 
def find_team_averages(team):
    #data frame is split up into teams
    #then find the performance of that team for the last x amount of games
    #.rolling and .mean pandas function is used to take the last x rows and get a meanof all the columns
    #you can change this around if you want to make predictions earlier in the season like for TRADES
    rolling = team.rolling(10).mean()
    #takes the current game and previous x-1 games 
    #finds the average performance of the team in the previous x amount of games
    return rolling

#this groups by team and season
#the average performance of the previous games is specific to each team
#grouping also by season because teams change a lot from season to season 
#only the respective season should be used
#roup_keys=False stops pandas from adding another level to the index
df_rolling = df_rolling.groupby(["team", "season"], group_keys=False).apply(find_team_averages)#applying find_team_averages function
```

```python
df_rolling
#NaN is returned from pandas because there are missing values 
#the missing values are because there aren't 10 previous games 
#this can be changed by the number of previous games that are looked at in find_team_averages
```

Combine rolling columns with regular columns

```python
#renaming the rolling columns so it doesnt overlap with the existing columns 
#_x indicates its the rolling average columns for the previous x games
rolling_cols = [f"{col}_x" for col in df_rolling.columns]
#rename columns by assigning to the .columns property 
df_rolling.columns = rolling_cols

#Concatonate the original data frame and the data frame with the rolling averages
#axis=1 combines the data frames using the columns to combine them it will make a wide data frame 
#axis=0 makes a tall data frame by putting the rows together
dataFrame = pd.concat([dataFrame, df_rolling], axis=1)
```

Improving accuracy- giving the algorithm some information we know in advance about the next game
In the real world we know if the next NBA basketball game will be at home or away and we know who the opponent will be
I will give the algorithm this information so it can make better judgements

When we call add_col it will run group by team because I want to get the next game home value for the specific team. Then it will apply the shift_col function and it will pass in the team data frame so data just for one team, finally i pass in the column name. 

The shift_col takes that column and shifts it one position by taking the next rows value and shift it back to the current row.

For example it will look at the home column find if the next game will be played home or away and put that in the current row.

This is also done for the columns team_next_opponent and date_next.

This is only done for information known in advance. For future games I know the team they will play against the date that they are going to play and if the team is home or away. I don't know for example the specific performance stats of the player or the team.

```python
#this function shift_colum takes in a team data frame and a column name
def shift_col(team, col_name):
    #creates a next column by taking the value from the next game and shift it back one row 
    next_col = team[col_name].shift(-1)
    return next_col

#function to add a column based on the shift_col function
#this will take in the main dataframe and the column name to aggregate based on 
def add_col(dataFrame, col_name):
    #group by teams 
    #applying the shift column function
    return dataFrame.groupby("team", group_keys=False).apply(lambda x: shift_col(x,col_name))

#using add_col function to give the algorithm information about the next game
#this is the information that everyone knows in advance about the games

#this column will indicate if the team will be home or away in its next game
#this creates a new column called home_next
#the output of this add column function will be assigned to the "home_next"
#passing in the data frame and the column "home"
dataFrame["home_next"] = add_col(dataFrame, "home")
#next opponent
dataFrame["team_opp_next"] = add_col(dataFrame, "team_opp")
#next date that they are going to play
dataFrame["date_next"] = add_col(dataFrame, "date")

#the copy warning can be a result of taking too many slices of a data frame and then you try to assign to it
#sometimes things will fail sometimes things will not
```
