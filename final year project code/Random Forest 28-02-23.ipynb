training Random Forest Model - this type of classifier can pick up non linear data - for example how the team codes dont show which team is better its just their assigned codes 

```python
from sklearn.ensemble import RandomForestClassifier 
```

n_estiomators the number of individual decision trees I want to train. Each tree in the series of trees having slightly different parameters, the higher n_estiomators number is the longer it will take to run but the potentially more accurate it will be. The higher  min_samples_split is the less likely we are to overfit but the lower the accuracy on training data will be. The games of GSW (Golden State Warriors) if the prediction is on the fourth game the model should look at the previous three games to help make better predictions. For example if the team is on a losing streak or a winning streak.

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    #closed = 'left' only have the rolling averages of the previous weeks not the week we are trying to calculate foe
    rolling_stats = group[cols].rolling(3,closed='left').mean()
    group[new_cols] = rolling_stats 
    #removes all of the rows that have missing values 
    group = group.dropna(subset=new_cols)
    return group
```

```python
cols = ["fg","3p","ft","ast","stl","blk","tov", "pts"]
new_cols = [f"{c}_rolling" for c in cols ]
```

```python
new_cols
```

This is just for the the GSW games 

```python
rolling_averages(group, cols, new_cols)
```

After getting the original nba_games dataframe and grouped it by teams which creates one data frame for each team in the data. then the lambda function was applied to compute the rolling averages.

```python
games_rolling = games.groupby("team").apply(lambda x:rolling_averages(x, cols, new_cols)) 
```
