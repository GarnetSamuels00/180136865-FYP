For a more complex data set it may need to be split multiple times to come up with the decision boundaries 
This is different in comparison to a simple data set where ridge regression can be used to draw when line for the decision boundaries
The order which the decsion tree is split up is important 
The order which the data sets attributes are selected will affect performance
To select the ordering system - 
You want an ordering to get a pure subset that gives the lowest entropy (the measure of randomness in the sample)
For example splitting the teams that average over a certain amount of points into groups. There may be a group that averages over a certain amount of points that always win their games. 1:3 or 6:2 win vs losses for the teams in hypothetical groups is relatively low entropy. This is low Entropy. Having a 50/50 split means there is a high entropy and the entopy is 1 total randomness. When teams are split up into groups based off of attributes you want the majority of the results to either be that the team wins or loses. 
With an ordering system that achieves low entropy there is a High Information Gain 
With an ordering system that achieves high entropy there is a Low Information Gain
Using an approach that leads to high information gain at every decision split 
Gini Impurity 
Where after splitting a sample ( putting a line at the midpoint for example there may be majority wins at the top and majority losses at the bottom but there might be one data point where ther is a win at the bottom) there wil be an anomalous data point. The decision split is almost pure but there is a little bit of impurity. Similar to entropy. Mathematics  


```python
import pandas as pd 
df = pd.read_csv("nba_games.csv", index_col=0)
df = df.reset_index(drop = True)
df = df.sort_values("date")
df = df.reset_index(drop = True)
df
```

```python
del df["mp.1"]
del df["mp_opp.1"]
del df["index_opp"]
```

```python
def get_target(team):
    team["target"] = team["won"].shift(-1)
    return team

df = df.groupby("team", group_keys=False).apply(get_target)
```
